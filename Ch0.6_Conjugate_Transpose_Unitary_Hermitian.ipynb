## 0.6 켤레전치·유니타리·에르미트 행렬

대거(†), 유니타리(게이트), 에르미트(관측량)를 검증하고, 심화로 스펙트럼 정리와 에르미트→유니타리($e^{-iHt}$) 관계를 확인한다. 게이트는 행렬 예시로만 다룬다.

### 0.6.1 켤레전치(대거): 전치 후 복소 켤레

```python
import numpy as np

A = np.array([[1+2j, 3],[0, 1-1j]])
print("A† =\n", A.conj().T)
```

### 0.6.2 유니타리 행렬: U†U = I (노름 보존)

```python
import numpy as np

X = np.array([[0,1],[1,0]])
H = (1/np.sqrt(2))*np.array([[1,1],[1,-1]])
for name,U in [("X",X),("H",H)]:
    assert np.allclose(U.conj().T @ U, np.eye(2))
print("검증 통과: X, H 모두 유니타리")
# 노름 보존 확인
psi = np.array([0.6, 0.8])
print("입력 노름:", np.linalg.norm(psi), "-> H 적용 후:", round(np.linalg.norm(H@psi),6))
```

### 0.6.3 에르미트 행렬: H† = H, 고윳값이 실수

```python
import numpy as np

Z = np.array([[1,0],[0,-1]])
print("Z† == Z ?", np.allclose(Z.conj().T, Z))
print("Z 고윳값:", np.linalg.eigvalsh(Z), "(실수)")
```

### 0.6.4 심화: 스펙트럼 정리와 측정
에르미트 행렬은 A = Σ λᵢ|i⟩⟨i| 로 분해된다(고윳값 실수, 고유벡터 정규직교). 또 에르미트 H에 대해 e^{-iHt}는 유니타리다.

```python
import numpy as np

# 스펙트럼 분해: A = Σ λ_i |i><i|
A = np.array([[2., 1.],[1., 2.]])
vals, vecs = np.linalg.eigh(A)   # 에르미트용
recon = sum(vals[i]*np.outer(vecs[:,i], vecs[:,i].conj()) for i in range(2))
print("재구성 A =\n", np.round(recon,6))
assert np.allclose(recon, A)
print("검증 통과: A = Σ λ_i |i><i|")
```

```python
import numpy as np
from scipy.linalg import expm

# 에르미트 H -> 유니타리 e^{-iHt}
H = np.array([[1,0],[0,-1]])   # 에르미트
t = 0.7
U = expm(-1j*H*t)
print("U = e^{-iHt} 가 유니타리?", np.allclose(U.conj().T @ U, np.eye(2)))
print("-> 관측량(에르미트)이 시간 진화(유니타리)를 생성한다")
```
