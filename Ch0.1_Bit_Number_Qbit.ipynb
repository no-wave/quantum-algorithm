## 0.1 비트·수 체계와 큐비트로 가는 길

고전 정보의 단위인 비트와 수 체계(2진·16진), 음수·실수의 비트 표현을 파이썬으로 다뤄 보고, 고전 비트와 큐비트의 차이를 직관적으로 확인한다.

### 0.1.1 비트와 수 체계
10진수는 각 자리에 10의 거듭제곱을 가중한 표현이다.

```python
# 10진수 215의 자리 전개: 5·10^0 + 1·10^1 + 2·10^2
digits = [5, 1, 2]   # 일의 자리부터
value = sum(d * 10**i for i, d in enumerate(digits))
print("자리 전개 합 =", value)   # 215
```

### 0.1.2 2진수와 16진수
컴퓨터는 0/1만 쓰므로 2진법을, 사람이 읽기 쉽게 4비트를 묶은 16진법을 함께 쓴다.

```python
# 10진수 ↔ 2진수 ↔ 16진수 변환
n = 215
print(f"10진수 {n}")
print(f"2진수  {bin(n)[2:]}")
print(f"16진수 {hex(n)[2:].upper()}")
# 2진수 1101101 을 10진수로 (2의 거듭제곱 합)
bits = "1101101"
val = sum(int(b) * 2**i for i, b in enumerate(reversed(bits)))
print(f"{bits}(2) = {val}(10)")
assert val == int(bits, 2)
print("검증 통과")
```

### 0.1.3 고전 비트에서 큐비트로
고전 비트는 0 또는 1 중 하나이지만, 큐비트는 0과 1의 중첩을 가진다.

```python
import numpy as np

classical_bit = 1
print("고전 비트:", classical_bit)
# 큐비트: |psi> = alpha|0> + beta|1> (균등 중첩 예시)
alpha = beta = 1/np.sqrt(2)
psi = np.array([alpha, beta])
print("큐비트 |psi> =", psi)
print("측정 확률 P(0), P(1) =", np.abs(psi)**2)
print("확률 합 =", np.sum(np.abs(psi)**2))   # 1
```

### 0.1.4 심화: 음수와 실수의 비트 표현
컴퓨터는 음의 정수를 **2의 보수**로, 실수를 **IEEE 754 부동소수점**으로 표현한다.

```python
# 2의 보수: 8비트로 음수 표현 (최상위 비트가 부호)
for x in [5, -5, -1, -128]:
    print(f"{x:>5} -> {format(x & 0xFF, '08b')}")
# -5 + 5 = 0 (같은 덧셈 회로로 처리됨)
print("(-5)+5 의 8비트:", format(((-5)&0xFF) + (5&0xFF) & 0xFF, '08b'))
```

```python
import struct

# IEEE 754 단정밀도(32비트): 부호1 + 지수8 + 가수23
x = 3.14
bits = ''.join(f'{b:08b}' for b in struct.pack('>f', x))
print(f"{x} 의 IEEE754 비트:")
print(f"  부호={bits[0]}  지수={bits[1:9]}  가수={bits[9:]}")
# 측정 비트열을 수로 해석하는 예: 2진 소수 0.101 -> 위상값
frac = "101"
phase = sum(int(b)*2**-(i+1) for i,b in enumerate(frac))
print(f"2진 소수 0.{frac} = {phase}  (QPE 위상 해석 방식)")
```
