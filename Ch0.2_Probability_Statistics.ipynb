## 0.2 확률과 통계 기초

확률 규칙(AND·OR)과 기술 통계(평균·분산·표준편차), 그리고 표본 분산 보정과 측정 표준오차를 다룬다.

### 0.2.1 확률의 기본 규칙

```python
# 독립 사건과 합집합
P_A, P_B = 0.3, 0.4
P_and = P_A * P_B                # 독립: P(A∩B)=P(A)P(B)
P_or  = P_A + P_B - P_and         # P(A∪B)=P(A)+P(B)-P(A∩B)
print(f"P(A∩B) = {P_and:.3f}")
print(f"P(A∪B) = {P_or:.3f}")
```

### 0.2.2 평균·분산·표준편차

```python
import numpy as np

data = np.array([15, 22, 33, 40, 55, 61, 79])
print(f"평균 μ     = {data.mean():.2f}")
print(f"분산 σ²    = {data.var():.2f}")
print(f"표준편차 σ = {data.std():.2f}")
assert np.isclose(data.var(), np.mean((data-data.mean())**2))
print("분산 수식 검증 통과")
```

### 0.2.3 측정 통계로의 연결
측정 횟수 N이 늘수록 확률 추정이 정밀해지며, 오차는 1/√N로 줄어든다.

```python
import numpy as np

np.random.seed(0); true_p = 0.5
for N in [10, 100, 1000, 10000]:
    est = (np.random.rand(N) < true_p).mean()
    print(f"N={N:>6}: 추정={est:.4f}, 오차={abs(est-true_p):.4f}")
```

### 0.2.4 심화: 표본 분산과 베셀 보정
표본으로 모분산을 추정할 때는 n 대신 n−1로 나눈다(베셀 보정). numpy는 `ddof`로 제어한다.

```python
import numpy as np

sample = np.array([15, 22, 33, 40, 55, 61, 79])
pop_var    = np.var(sample, ddof=0)   # 모분산(n으로 나눔)
sample_var = np.var(sample, ddof=1)   # 표본분산(n-1로 나눔, 불편추정량)
print(f"모분산(ddof=0)   = {pop_var:.2f}")
print(f"표본분산(ddof=1) = {sample_var:.2f}")
print("표본분산이 더 큼:", sample_var > pop_var)
```

```python
import numpy as np

# 측정 추정 확률의 표준오차: sqrt(p(1-p)/N)
p = 0.5
for N in [100, 1000, 10000]:
    se = np.sqrt(p*(1-p)/N)
    print(f"N={N:>6}: 표준오차 = {se:.4f}")
print("정밀도 10배 -> 측정 100배 필요 (1/√N)")
```
