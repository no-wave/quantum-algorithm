## 0.5 텐서곱

`np.kron`으로 텐서곱을 계산하고, 다중 큐비트 상태·차원 폭증을 본 뒤, 심화로 곱 상태와 얽힘을 슈미트 분해로 구분한다.

### 0.5.1 텐서곱의 정의

```python
import numpy as np

a = np.array([1,2]); b = np.array([3,5])
print("a ⊗ b =", np.kron(a,b))   # [3,5,6,10]
```

### 0.5.2 다중 큐비트 상태

```python
import numpy as np

ket0 = np.array([1,0]); ket1 = np.array([0,1])
print("|0>⊗|1> = |01> =", np.kron(ket0,ket1))   # [0,1,0,0]
I = np.eye(2); X = np.array([[0,1],[1,0]])
print("I⊗X 크기:", np.kron(I,X).shape)
```

### 0.5.3 지수적 차원 폭증

```python
for n in [1,2,3,5,10]:
    print(f"{n:>2} 큐비트 -> {2**n:>5} 차원")
print("50 큐비트 -> 2^50 ≈ {:.2e} 차원".format(2.0**50))
```

### 0.5.4 심화: 곱 상태와 얽힘, 슈미트 분해
텐서곱으로 분해되면 곱 상태, 안 되면 얽힘 상태다. 슈미트 분해(=SVD)의 0이 아닌 항 수(슈미트 랭크)로 판별한다.

```python
import numpy as np

def schmidt_rank(state):
    # 2큐비트 상태(4차원)를 2x2 행렬로 재배열한 뒤 특이값으로 랭크 계산
    M = state.reshape(2,2)
    s = np.linalg.svd(M, compute_uv=False)
    return int(np.sum(s > 1e-10)), np.round(s,4)

# 곱 상태: |0>⊗|+>
plus = np.array([1,1])/np.sqrt(2)
prod = np.kron(np.array([1,0]), plus)
# 얽힘 상태: 벨 상태 (|00>+|11>)/√2
bell = np.array([1,0,0,1])/np.sqrt(2)
for name, st in [("곱 상태 |0>|+>", prod), ("벨 상태", bell)]:
    r, s = schmidt_rank(st)
    print(f"{name}: 슈미트 랭크={r}, 특이값={s} -> {'얽힘' if r>1 else '곱 상태'}")
```
