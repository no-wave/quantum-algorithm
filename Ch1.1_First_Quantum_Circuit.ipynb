## 1.1 기본 양자 회로 (Basic Quantum Circuit)

```python
%pip install -q qiskit qiskit-aer
```

```python
# 필요한 라이브러리 불러오기
from qiskit import QuantumCircuit, transpile        # 양자 회로 생성 / 백엔드용 변환(트랜스파일)
from qiskit_aer import AerSimulator                  # 양자 회로를 흉내 내는 고전 시뮬레이터
from qiskit.visualization import plot_histogram      # 측정 결과 막대그래프
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 큐비트 1개 + 고전 비트 1개로 양자 회로 생성
qc = QuantumCircuit(1, 1)
```

```python
# 큐비트를 중첩 상태로 만들기 위해 하다마드 게이트 적용
qc.h(0)
```

```python
# 큐비트 측정
qc.measure(0, 0)
```

```python
# Qiskit Aer 시뮬레이터로 회로 실행
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
result = simulator.run(compiled_circuit).result()
```

```python
# 측정 결과(각 값이 나온 횟수) 집계
counts = result.get_counts()
print("Counts:", counts)
```

```python
# 회로를 텍스트로 출력
print(qc)
```

```python
# 회로를 그림으로 시각화
qc.draw('mpl')
plt.show()
```

```python
# 측정 결과 시각화
plot_histogram(counts)
plt.show()
```
